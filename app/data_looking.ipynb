```python
import pandas as pd
import sqlite3
import json
```

```python
file_path = "../data/events.jsonl"

target_user = "031ddc37-6878-4af6-b043-09a5ad23b653"

match_start_count_file = 0

with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        try:
            row = json.loads(line)
        except json.JSONDecodeError:
            continue

        if (
            row.get("event_type") == "match_start"
            and row.get("user_id") == target_user
        ):
            match_start_count_file += 1

print("Match starts in JSONL file:", match_start_count_file)
```

```python
db_path = "../nord_challenge.db"  # adjust if needed

conn = sqlite3.connect(db_path)
cursor = conn.cursor()

cursor.execute("""
    SELECT COUNT(*) 
    FROM match_events
    WHERE event_type = 'match_start' AND user_id = '031ddc37-6878-4af6-b043-09a5ad23b653'
""")

match_start_count_db = cursor.fetchone()[0]

print("Match starts in DB:", match_start_count_db)

conn.close()
```
